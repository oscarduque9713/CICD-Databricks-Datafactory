```python
dbutils.widgets.removeAll()
```

```python
%sql
create widget text storageName default "adlssmartprojectdev13";
create widget text catalogo default "project_sedev";
create widget text container default "semiestrcturprojectdev";
```

```python
container = dbutils.widgets.get("container")
catalogo = dbutils.widgets.get("catalogo")
storageName = dbutils.widgets.get("storageName")

# definir rutas
path_inicio = f"abfss://{container}@{storageName}.dfs.core.windows.net/raw/"
path_target = f"{catalogo}.bronze."
```

```python
# importar functions essentials
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

## Leer JSON

```python
df_instances_train = spark.read.json(f"{path_inicio}instances_train2017.json")
```

```python
df_instances_train.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(f"{path_target}instances_train")
```
